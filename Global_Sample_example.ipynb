```python
import geopandas
from ast import literal_eval
from shapely import wkt
import folium
import pandas as pd
```

```python
!python3 Global_Sample.py 10
```

```python
world_df = pd.read_csv('../data/Global_samples_10.csv', index_col=[0])
world_df.head()
world_df['geometry'] = world_df['geometry'].apply(wkt.loads)
world_df['samples'] = world_df['samples'].apply(literal_eval)
world_gdf = geopandas.GeoDataFrame(world_df, geometry='geometry')
world_gdf.head()
```

```python
sample_map = folium.Map(location=[30, 30], zoom_start=2)
random_pixels = [element for sublist in world_gdf['samples'].tolist() for element in sublist]
random_pixels2 = [(lat, lon) for lon, lat in random_pixels]
for i in range(len(random_pixels2)):
    folium.Marker(random_pixels2[i], tooltip=str(random_pixels2[i])).add_to(sample_map)
sample_map
```
